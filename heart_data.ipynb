```python
import pandas as pd
data=pd.read_csv("heart.csv")
data
```

```python
data.info()
```

```python
missingvalue=data.isnull()
print(missingvalue)
```

```python
m=data.isnull().sum()
print(m)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data["SEX"]=label_encoder.fit_transform(data["Sex"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data["Chest_Pain_Type"]=label_encoder.fit_transform(data["ChestPainType"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data["St_Slope"]=label_encoder.fit_transform(data["ST_Slope"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data["Resting_ECG"]=label_encoder.fit_transform(data["RestingECG"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data["Exercise_Angina"]=label_encoder.fit_transform(data["ExerciseAngina"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data["Old_peak"]=label_encoder.fit_transform(data["Oldpeak"])
```

```python
x=data.drop(["Sex","ChestPainType","RestingECG","ExerciseAngina","Oldpeak","ST_Slope","HeartDisease"],axis=1)
```

```python
x.head()
```

```python
y=data["HeartDisease"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
data.info()
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
x_test
```

```python
y_test
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
import joblib 
joblib.dump(model,'heart1.pk1')
```
